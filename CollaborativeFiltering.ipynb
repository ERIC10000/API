```python
# Import Datasets
# Movie Recommendation System

# Users: How many user have viewed a particular movie
import pandas
users = pandas.read_csv("https://modcom.co.ke/datasets/file.csv")
users.head(5)
```

```python
# Movie Details
movies = pandas.read_csv("https://modcom.co.ke/datasets/Movie_Id_Titles.csv")
movies.head(5)
```

```python
# Movie Details for particular movie_id(item_id)
# Movie Title for item_id 50(n-1)
# loc

movies.loc[49]
movies.loc[171]
```

```python
# Merge users and movies in one dataframe(merge -> on the common column)
data = pandas.merge(users, movies, on='item_id')
data.head(5)
```

```python
# Data Analysis
# 1. Maximum Rating Per Movies
# Ans -> 5

data['rating'].max()
```

```python
# 2. Min Rating
# Ans -> 1
data['rating'].min()
```

```python
# 3. Average Movie Rating
data['rating'].mean()
```

```python
# Statistics
# describe()
data.describe()
```

```python
# Correlation
data.corr()
```

```python
# Modelling -> Design Recommender System
# Based:
# 1. Ratings of the Movies(Highest rated movies)
# 2. Viewers(How Many users viewed the movie -> Highest number of Viewers)
```

```python
# 1.Group Movies based the Average Rating
title_mean_ratings = data.groupby('title')['rating'].mean()
# Which were the highest rated movies? -> Sort values in descending manner
title_mean_ratings.sort_values(ascending=False).head(15)
```

```python
# 2. How Many Users Rated a Particular Movie Item
# Assumption: Higher the number of users the better the recommendation
# Which movie had the highest viewers?
user_rating_count = data.groupby('title')['rating'].count()
user_rating_count.sort_values(ascending=False).head(15)
```

```python
# Display the Mean Ratings and User Counts on a DataFrame
mean_rating_per_movie = pandas.DataFrame(title_mean_ratings)
mean_rating_per_movie['number_of_views'] = pandas.DataFrame(user_rating_count)
mean_rating_per_movie.sort_values('number_of_views', ascending=False).head(15)
```

```python
# Pivot Table: Pool to Show, each movie title, number of rating, the user
# All the Movies, All the Users, All the Titles
pivot = data.pivot_table(index = 'user_id', columns = 'title', values = 'rating')
pivot
```

```python
# Movie Selection
selected_movie = pivot['101 Dalmatians (1996)']
```

```python
# Getting Similar Movie based on rating and Viewers
similar = pivot.corrwith(selected_movie)
```
