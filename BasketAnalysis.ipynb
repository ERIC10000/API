```python
# WIFI : MODCOM2023
# password: Modcom#20

# BASKET ANALYSIS:
import pandas
data = pandas.read_csv("https://modcom.co.ke/data/datasets/store.csv")
data.head()
```

```python
# Regions:Different Regions have different taste and preferences
data.groupby('Country').size().sort_values(ascending=False)
```

```python
data = data[data['Country'] == 'Netherlands']
data.head()
# Adjusting to the region
```

```python
# Checking for invoice no and description and showing the total product quantity
data.groupby(['InvoiceNo', 'Description'])['Quantity'].sum()
```

```python
# Pivot Table
# index = InvoiceNo, columns = 'Description', values= 'Quantity'
# pandas.set_option('display.max_columns', None)
# pandas.set_option('display.max_rows', None)
basket = data.pivot_table(index='InvoiceNo', columns='Description', values = 'Quantity', fill_value=0)
basket
```

```python
# A function: Check if a quantity > 0:(1) else(0)

def convert(quantity):
    if quantity > 0:
        return 1
    else:
        return 0
```

```python
# Apply the function to each item quantity on the basket
basket_set = basket.applymap(convert)
basket_set
```

```python
from mlxtend import frequent_patterns
# Apply the Apriori Algorithm with Association Rules:
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

# consequent
# antecedent

# If A(antecedent) then B(consequent)

frequent_items_sets = apriori(basket_set, min_support= 0.10, use_colnames=True)
rules = association_rules(frequent_items_sets)

rules
```
